### Concept 1: Coefficient of Determination（R-squared）

Let's go ahead and see equation for the R-squared value

The coefficient of determination is a number between 0 and 1 that measures how well a statistical model predicts an outcome.

#### Simple Example

2. Step 2 find RSS: the sum of the residuals squared

The larger the residuals squared meaning the further off the data set is.
