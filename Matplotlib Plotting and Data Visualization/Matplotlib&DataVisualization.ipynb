### Pyplot API vs. Object-Oriented API and Pyplot life cycle

there are two type of matplotlib - the pyplot api and the object-oriented API. 
The pyplot API is less-flexible. 


```python
# first method

fig = plt.figure() # creates a figure
ax = fig.add_subplot() # adds sp,e axes

plt.show()
```

```python
# 2nd method

fig = plt.figure() #creates a figure
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)# add some data
plt.show()
```

```python
# third method(recommended)

fig,ax = plt.subplots()
ax.plot(x,y); #add some data
```

notice that every time we use plot, we refresh the data.

## Matplotlib example workflow

```python
#### 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Set up plot
fig,ax = plt.subplots(figsize = (10,10)) # dimention of the plot is 10 w x 10h

# 3. plot data - add data to axis
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "simple plot",
       xlabel = "x-axis",
       ylabel =  "y-axis"
       )

# 5. save and show(you save the whole figure)
fig.savefig("./images/sample-plot.png")
```

### Two options for subplots

```python
# Subplot Option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                        ncols = 2,
                                        figsize = (10,5))
ax1.plot(x,x/2);
ax2.hist(x);
ax3.scatter(np.random.random(10),np.random.random(10));
ax4.bar(nut_butter_prices.keys(),nut_butter_prices.values());
```

```python
# subplot option 2

fig,ax = plt.subplots(nrows = 2,ncols = 2,figsize=(10,5))
ax[0,0].plot(x,x/3);# plot the row 0, col 0 ax
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());

```

### Creating OO plot from scratch

```python
#create somesort of data: explore the hear disease for someone older than 80

over_50 = heart_disease[heart_disease["age"]>50]
len(over_50)
over_50.head()
```

```python
#using pyplot method -- not ideal
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target')
```

```python
## OO method of Matplotlib with Pyplot method

fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax); 
# the last item means that we want the ax of the plot be the ax
# created using subplots

#ax.set_xlim([45,100]) # set x range to be 45 - 100
```

```python
# try creating a plot using completly just OO method. 

fig,ax = plt.subplots(figsize = (10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                     y = over_50["chol"],
                     c = over_50["target"])

#customize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = 'Age',
       ylabel = 'Cholesterol')
# add a legend

ax.legend(*scatter.legend_elements(),title = "Target") # the * means to unpack the element
# *scatter -> will go into the scatter variables.
# *scatter.legend_element() -> will find the legend_elements, in this case, the "target", 
# then it will find the value of targe and color it. 

# more customization-- adding a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = "--")
```

### more customizeation using matplotlib

```python
#subplot of chol, age, thalach
#step 1 of using OO to plot
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize = (10,10))

# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y =over_50["chol"],
                      c = over_50["target"])

# customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel = "Age",
        ylabel = "CHolesterol")
#add a legend to ax0

ax0.legend(*scatter.legend_elements(),title = "Target")

ax0.axhline(y=over_50["chol"].mean(),
            linestyle = '--')

scatter = ax1.scatter(x=over_50["age"],
                      y =over_50["thalach"],
                      c = over_50["target"])


ax1.legend(*scatter.legend_elements(),title = "Target")

ax1.axhline(y=over_50["thalach"].mean(),
            linestyle = '--')

fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight = 'bold')
```
